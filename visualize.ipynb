```python
import spacy
from spacy import displacy
from spacy.gold import GoldCorpus
from spacy.gold import offsets_from_biluo_tags
nlp = spacy.load("en_core_web_sm")
```

```python
corpus = GoldCorpus('/mnt/DATA/data/nlp/FIN5.json', '/mnt/DATA/data/nlp/FIN5.json')
data = corpus.train_docs(nlp)
doc, gold = next(data)
```

```python
spacy_ner = lambda doc, gold: {'text': doc.text, 
                               'ents': [{'start': ent[0], 'end': ent[1], 'label': ent[2]} 
                                        for ent in offsets_from_biluo_tags(doc, gold.ner)]}
```

```python
displacy.render(spacy_ner(doc, gold), style="ent", manual=True)
```

```python
doc, gold = next(data)
displacy.render(spacy_ner(doc, gold), style="ent", manual=True)
```

```python
doc, gold = next(data)
displacy.render(spacy_ner(doc, gold), style="ent", manual=True)
```
